```python
from matplotlib import pyplot as plt

plt.rc('axes', unicode_minus = False)
plt.rc('font',family='Malgun Gothic')

#plt.rcParams['axes.unicode-minus'] = False
#plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
x = [0, 1, 2, 3]
y = [1, 4, 9, 16]

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.xlabel("X축")
plt.ylabel("Y축")
plt.grid()
```
